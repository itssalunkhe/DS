```python
def data_gender(x):
    if x == 'M':
        return 0
    else :
        return 1
```

```python
def data_ssc_b(x):
    if x == 'Others':
        return 0
    else:
        return 1
```

```python
def data_hsc_b(x):
    if x == 'Others':
        return 0
    else:
        return 1
```

```python
def data_hsc_s(x):
    if x == 'Commerce':
        return 1
    elif x == 'Science':
        return 2
    else :
        return 3
```

```python
def data_degree_t(x):
    if x == 'Sci&Tech':
        return 1
    elif x == 'Comm&Mgmt':
        return 2
    else :
        return 3
```

```python
def data_workex(x):
    if x == 'No':
        return 0
    else:
        return 1
```

```python
def data_specialisation(x):
    if x == 'Mkt&HR':
        return 0
    else:
        return 1
```

```python
def data_status(x):
    if x == 'Placed':
        return 0
    else:
        return 1
```

```python
df['gender'] = df['gender'].apply(data_gender)
df['ssc_b'] = df['ssc_b'].apply(data_ssc_b)
df['hsc_b'] = df['hsc_b'].apply(data_hsc_b)
df['hsc_s'] = df['hsc_s'].apply(data_hsc_s)
df['degree_t'] = df['degree_t'].apply(data_degree_t)
df['workex'] = df['workex'].apply(data_workex)
df['specialisation'] = df['specialisation'].apply(data_specialisation)
df['status'] = df['status'].apply(data_status)
```

```python
df = df.fillna({'salary' : 0})
```

```python
df['salary'].unique()
```

```python
df1=df.astype(int)
```

```python
X = df1.drop('salary',axis=1)
Y = df1['salary']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
```

```python
sgd = linear_model.SGDClassifier(max_iter=5, tol=None)
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)

sgd.score(X_train, Y_train)

acc_sgd = round(sgd.score(X_test, Y_test) * 100, 2)
```

```python
linreg = LinearRegression()
linreg.fit(X_train,Y_train)
Y_pred = linreg.predict(X_test)

linreg.score(X_train, Y_train)


acc_linreg = round(linreg.score(X_test, Y_test) * 100, 2)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)

Y_pred = logreg.predict(X_test)

acc_log = round(logreg.score(X_test, Y_test) * 100, 2)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3) 
knn.fit(X_train, Y_train)  
Y_pred = knn.predict(X_test)  
acc_knn = round(knn.score(X_test, Y_test) * 100, 2)
```

```python
gaussian = GaussianNB() 
gaussian.fit(X_train, Y_train)  
Y_pred = gaussian.predict(X_test)  
acc_gaussian = round(gaussian.score(X_test, Y_test) * 100, 2)
```

```python
perceptron = Perceptron(max_iter=5)
perceptron.fit(X_train, Y_train)

Y_pred = perceptron.predict(X_test)

acc_perceptron = round(perceptron.score(X_test, Y_test) * 100, 2)
```

```python
linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)

Y_pred = linear_svc.predict(X_test)

acc_linear_svc = round(linear_svc.score(X_test, Y_test) * 100, 2)
```

```python
decision_tree = DecisionTreeClassifier() 
decision_tree.fit(X_train, Y_train)  
Y_pred = decision_tree.predict(X_test)  
acc_decision_tree = round(decision_tree.score(X_test, Y_test) * 100, 2)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)

Y_prediction = random_forest.predict(X_test)

random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_test, Y_test) * 100, 2)
```

```python
results = pd.DataFrame({
    'Model': ['Linear Regression','Logistic Regression','KNN','Support Vector Machines',   
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 
              'Decision Tree'],
    'Score': [acc_linreg, acc_log, acc_knn, acc_linear_svc,  
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_decision_tree]})
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
result_df.head(9)
```
