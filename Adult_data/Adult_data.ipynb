```python
df2 = df.rename(columns={'education-num' : 'education_num',
                 'marital-status' : 'marital_status',
                 'native-country' : 'native_country'})
df2.head()
```

```python
dumm_workclass = pd.get_dummies(df2.workclass)
dumm_workclass1 = dumm_workclass.drop('?',axis=1)

dumm_education = pd.get_dummies(df2.education)
dumm_education1 = dumm_education.drop('Masters',axis=1)

dumm_marital_status = pd.get_dummies(df2.marital_status)
dumm_marital_status1 = dumm_marital_status.drop('Divorced',axis=1)

dumm_occupation = pd.get_dummies(df2.occupation)
dumm_occupation1 = dumm_occupation.drop('?',axis=1)

dumm_relationship = pd.get_dummies(df2.relationship)
dumm_relationship1 = dumm_relationship.drop('Unmarried',axis=1)

dumm_race = pd.get_dummies(df2.race)
dumm_race1 = dumm_race.drop('Other',axis=1)

dumm_native_country = pd.get_dummies(df2.native_country)
dumm_native_country1 = dumm_native_country.drop('?',axis=1)
```

```python
df3 = pd.concat([df2.drop(['workclass','education','marital_status','occupation','relationship','race','native_country'],axis=1),
                 dumm_workclass1,dumm_education1,dumm_marital_status1,
                 dumm_occupation1,dumm_relationship1,dumm_race1,dumm_native_country1],
                 axis=1)
```

```python
df3.head()
```

```python
df3['salary'].unique()
```

```python
def data_salary(x):
    if x == '<=50K' :
        return 0
    else :
        return 1
```

```python
def data_sex(x):
    if x =='Male' :
        return 0
    else :
        return 1
```

```python
df3['salary'] = df3['salary'].apply(data_salary)
df3['sex'] = df3['sex'].apply(data_sex)
```

```python
df3.head()
```

```python
df3['salary'].unique()
```

```python
X = df3.drop('salary',axis=1)
Y = df3['salary']
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
```

```python
sgd = linear_model.SGDClassifier(max_iter=5, tol=None)
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)

sgd.score(X_train, Y_train)

acc_sgd = round(sgd.score(X_test, Y_test) * 100, 2)
```

```python
linreg = LinearRegression()
linreg.fit(X_train,Y_train)
Y_pred = linreg.predict(X_test)

linreg.score(X_train, Y_train)


acc_linreg = round(linreg.score(X_test, Y_test) * 100, 2)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)

Y_pred = logreg.predict(X_test)

acc_log = round(logreg.score(X_test, Y_test) * 100, 2)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3) 
knn.fit(X_train, Y_train)  
Y_pred = knn.predict(X_test)  
acc_knn = round(knn.score(X_test, Y_test) * 100, 2)
```

```python
gaussian = GaussianNB() 
gaussian.fit(X_train, Y_train)  
Y_pred = gaussian.predict(X_test)  
acc_gaussian = round(gaussian.score(X_test, Y_test) * 100, 2)
```

```python
perceptron = Perceptron(max_iter=5)
perceptron.fit(X_train, Y_train)

Y_pred = perceptron.predict(X_test)

acc_perceptron = round(perceptron.score(X_test, Y_test) * 100, 2)
```

```python
linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)

Y_pred = linear_svc.predict(X_test)

acc_linear_svc = round(linear_svc.score(X_test, Y_test) * 100, 2)
```

```python
decision_tree = DecisionTreeClassifier() 
decision_tree.fit(X_train, Y_train)  
Y_pred = decision_tree.predict(X_test)  
acc_decision_tree = round(decision_tree.score(X_test, Y_test) * 100, 2)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)

Y_prediction = random_forest.predict(X_test)

random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_test, Y_test) * 100, 2)
```

```python
results = pd.DataFrame({
    'Model': ['Linear Regression','Logistic Regression','KNN','Support Vector Machines',   
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 
              'Decision Tree'],
    'Score': [acc_linreg, acc_log, acc_knn, acc_linear_svc,  
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_decision_tree]})
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
result_df.head(9)
```
